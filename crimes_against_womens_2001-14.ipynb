# Data Visualization

# 1. Total number of crimes individually

```python

total_rapes = df['Rape'].sum()
total_Kidnapping_Abduction= df['Kidnapping and Abduction'].sum()
total_dowry_deaths= df['Dowry Deaths'].sum()
total_assault_case= df['Assault on women with intent to outrage her modesty'].sum()
total_insult_case= df['Insult to modesty of Women'].sum()
total_cruelty_case= df['Cruelty by Husband or his Relatives'].sum()
total_girls_import_case= df['Importation of Girls'].sum()

print(f"Total number of Rape cases: {total_rapes}")
print(f"Total number of Kidnapping & Abduction cases: {total_Kidnapping_Abduction}")
print(f"Total number of Dowry death cases: {total_dowry_deaths}")
print(f"Total number of Assault cases: {total_assault_case}")
print(f"Total number of Women insult cases: {total_insult_case}")
print(f"Total number of Cruelty cases: {total_cruelty_case}")
print(f"Total number of Girls importing cases: {total_girls_import_case}")
```

# 2. Crime Category Distribution

```python
data = {
    'Crime Type': ['Rape', 'Kidnapping & Abduction', 'Dowry death', 'Assault', 'Women insult', 'Cruelty', 'Girls importing'],
    'Total Cases': [619158, 746198, 215480, 1212258, 292756, 2233888, 1872]
}

df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
plt.yscale('log')

plt.bar(df['Crime Type'], df['Total Cases'], color='skyblue')
plt.xlabel('Crime Type')
plt.ylabel('Total Cases')
plt.title('Total Number of Womens Cases by Crime Type')
plt.xticks(rotation=45, ha='right') 

plt.tight_layout()
plt.show()
```

 # Conclusion: Most of the women suffers from Assault and Cruelty, even we can see the cases of Rape and Kidnapping are also greater.

# 3. Total crime over time

# 4. The top 10 Unnsafe states with the highest number of crimes reported from 2001-2014

```python
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']  
df['total_crimes'] = df[crime_columns].sum(axis=1)
statewise_crime_distribution = df.pivot_table(index='STATE/UT', values='total_crimes', aggfunc='sum')
statewise_crime_distribution_sorted = statewise_crime_distribution.sort_values(by='total_crimes', ascending=False)
plt.figure(figsize=(16, 6))
statewise_crime_distribution_sorted.head(10).plot(kind='bar', color= 'plum')
plt.title('Top 10 Unsafe States/UTs with Highest Crimes (2001-2014)', fontsize=16)
plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Total Crimes', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion: States like A.P, U.P, W.B, M.P & Rajasthan are showing top 5 unsafe states for womens followed by Maharashtra, Assam, Kerala, Gujrat & Bihar.

# 5. The top 10 safe states with the lowest crimes reported from 2001-2014

```python
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']  
df['total_crimes'] = df[crime_columns].sum(axis=1)
statewise_crime_distribution = df.pivot_table(index='STATE/UT', values='total_crimes', aggfunc='sum')
statewise_crime_distribution_sorted = statewise_crime_distribution.sort_values(by='total_crimes', ascending=False)
plt.figure(figsize=(16, 6))
statewise_crime_distribution_sorted.tail(10).plot(kind='bar', color= 'lightgreen')
plt.title('Top 10 Safe States/UTs with Lowest Crimes (2001-2014)', fontsize=16)
plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Total Crimes', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion: States like sikkim, Puducherry, Nagaland, Daman & Diu etc are showing safe States and UT'S.

# 6. Yearly Crime Trend by Category

```python

crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']  
df_melted = df.melt(id_vars=['Year'], value_vars=crime_columns, var_name='Crime Category', value_name='Count')

# Plotting the yearly crime trend by category
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x='Year', y='Count', hue='Crime Category', marker='o')
plt.title('Yearly Crime Trend by Category', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Crime Category')
plt.tight_layout()
plt.show()
```

# Conclusion: It highlights an increase in crimes like cruelty by husbands and assault on women, with a steady upward trend over time.

# 7. Change in Crime Rates (2001-2014)

```python
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']  
crime_trend_by_year_category = df.groupby('Year')[crime_columns].sum().reset_index()
plt.figure(figsize=(12, 8))
for crime in crime_columns:
    sns.lineplot(data=crime_trend_by_year_category, x='Year', y=crime, marker='o', label=crime)

plt.title('Change in Crime Rates by Category (2001-2014)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.legend(title='Crime Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Conclusion: This graph depicts the change in crime rates by category from 2001 to 2014, showing a steady rise in crimes like cruelty by husbands and assault on women. The y-axis represents the number of crimes, and it highlights significant increases in kidnapping, rape, and assault cases over time.
